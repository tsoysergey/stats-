#### Formula to calculate the Z-score z = (X-mu) / SD
#### Our condition was "no more than 6%". Thus, 0.5 - 0.06 = 0.44. We look at the Z-score table and find that there are two suitable values 1.55 and 1.56. I would go for 1.56.

#### right now, we have z = -1.56, mean = 63558 and standard deviation = 2546
#### Plugging in these values, we find that X = 59586

```python
minimum_guaranteed_mileage = 59586
```

```python
# funtion to draw corresponding Z-value
def draw_z_score(x, cond, mu, sigma, title):
    y = norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, norm.pdf(z, mu, sigma))
    plt.title(title)
    plt.show()
```

```python
x = np.arange(50000, 75000,500)
minimum_guaranteed_mileage = 59586
draw_z_score(x, x<minimum_guaranteed_mileage, mean_value,std_value, 'Normal Distribution')
```

### So, now we can advertise that those tires that wears out before 59586 miles will be replaced for free.
